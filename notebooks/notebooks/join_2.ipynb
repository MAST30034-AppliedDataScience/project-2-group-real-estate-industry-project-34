```python
import pandas as pd

model_df = pd.read_parquet("../data/curated/model_data.parquet")
```

```python
# Step 1: Read the CSV file into a DataFrame
csv_file_path = '../data/curated/listings_amenity_final.csv'  # Replace with your CSV file path
df = pd.read_csv(csv_file_path)

# Step 2: Convert the DataFrame to a Parquet file
parquet_file_path = '../data/curated/listings_amenity_final.parquet'  # Specify the desired Parquet file path
df.to_parquet(parquet_file_path, index=False)

# Step 3: Read the Parquet file back into a DataFrame
distance_df = pd.read_parquet(parquet_file_path)
```

```python
distance_df[['latitude', 'longitude']].drop_duplicates(inplace=True)
```

```python

# Filter columns in distance_df that start with 'distance_to'
distance_columns = [col for col in distance_df.columns if col.startswith('distance_to')]

# Merge distance data with model_df based on 'SA2_CODE21' or other matching criteria
model_df = pd.merge(model_df, distance_df[['latitude', 'longitude'] + distance_columns], 
                    on=['latitude', 'longitude'], how='left')
```

```python
nan_count = model_df['distance_to_mall'].isna().sum()
print(nan_count)
```
