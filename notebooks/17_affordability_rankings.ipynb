This notebook uses median income data obtained from ABS to rank affordability in SA2s based on the definition that an affordable property is one where its rental price is 30% of less than the median rental income. 

```python
import pandas as pd
```

```python
predicted_incomes = pd.read_csv('/Users/danielletran/Desktop/ADS_A2/project-2-group-real-estate-industry-project-34/data/landing/median_income_predictions-2011-2027-final.csv')
current_rentals = pd.read_parquet('/Users/danielletran/Desktop/ADS_A2/project-2-group-real-estate-industry-project-34/data/landing/model_data_with_income.parquet')
predicted_rentals = pd.read_csv('/Users/danielletran/Desktop/ADS_A2/project-2-group-real-estate-industry-project-34/data/landing/sa2_growth_filtered.csv')
```

```python
current_rentals = current_rentals[current_rentals['year'] == 2024]
```

```python
current_rentals
```

```python
def individual_rent(persons, rent):
    return rent/persons
```

```python
current_rentals['price'] = current_rentals.apply(lambda row: individual_rent(row['bedrooms'], row['price']), axis = 1) 
```

```python
current_rentals = current_rentals.groupby('SA2_CODE21')['price'].median().reset_index()
current_rentals = current_rentals.rename(columns={'price': 'predicted_price_2024'})
```

```python
predicted_rentals
```

```python
predicted_incomes
```

```python
def get_30(yearly):
    weekly = yearly / 48 #52 weeks in a year
    proportion = weekly * 0.3

    return proportion
```

```python
predicted_incomes_filtered = predicted_incomes[['Statistical Areas Level 2 2021 code','2024_Forecast','2025_Forecast', '2026_Forecast', '2027_Forecast']]
predicted_incomes_filtered = predicted_incomes_filtered.rename(columns={'Statistical Areas Level 2 2021 code': 'SA2_CODE21'})
```

```python
predicted_incomes_filtered[['2024_Forecast','2025_Forecast', '2026_Forecast', '2027_Forecast']] = predicted_incomes_filtered[['2024_Forecast','2025_Forecast', '2026_Forecast', '2027_Forecast']].apply(get_30)
```

```python
predicted_incomes_filtered
```

```python
predicted_rentals[['SA2_CODE21','predicted_price_2025', 'predicted_price_2026','predicted_price_2027']]
```

```python
# Convert SA2_CODE21 to integer in both DataFrames before merging
predicted_incomes_filtered['SA2_CODE21'] = predicted_incomes_filtered['SA2_CODE21'].astype(int)
predicted_rentals['SA2_CODE21'] = predicted_rentals['SA2_CODE21'].astype(int)
current_rentals['SA2_CODE21'] = current_rentals['SA2_CODE21'].astype(int)

# Now merge the DataFrames
combined_prediction = predicted_incomes_filtered.merge(predicted_rentals, on='SA2_CODE21')
combined_prediction = combined_prediction.merge(current_rentals, on='SA2_CODE21')

# Display the merged DataFrame
print(combined_prediction.head())
```

```python
combined_prediction = predicted_incomes_filtered.merge(predicted_rentals,on='SA2_CODE21')
combined_prediction = combined_prediction.merge(current_rentals, on = 'SA2_CODE21')
```

```python
combined_prediction
```

```python
def calculate_affordability(weekly_rent, weekly_income_proportion):
    difference = weekly_rent - weekly_income_proportion

    #if difference is positive, rent is more than 30% of the income

    #if difference is negative, rent is less than 30% of the income

    return difference
```

```python
combined_prediction['2024_affordability_difference'] = combined_prediction.apply(
    lambda row: calculate_affordability(row['predicted_price_2024'], row['2024_Forecast']), axis=1)

combined_prediction['2025_affordability_difference'] = combined_prediction.apply(
    lambda row: calculate_affordability(row['predicted_price_2025'], row['2025_Forecast']), axis=1)

combined_prediction['2026_affordability_difference'] = combined_prediction.apply(
    lambda row: calculate_affordability(row['predicted_price_2026'], row['2026_Forecast']), axis=1)

combined_prediction['2027_affordability_difference'] = combined_prediction.apply(
    lambda row: calculate_affordability(row['predicted_price_2027'], row['2027_Forecast']), axis=1)
```

```python
tfouraffordability = combined_prediction[['SA2_CODE21','Suburb','2024_affordability_difference']].sort_values(by='2024_affordability_difference')
tfiveaffordability = combined_prediction[['SA2_CODE21','Suburb','2025_affordability_difference']].sort_values(by='2025_affordability_difference')
tsixaffordability = combined_prediction[['SA2_CODE21','Suburb','2026_affordability_difference']].sort_values(by='2026_affordability_difference')
tsevenaffordability = combined_prediction[['SA2_CODE21','Suburb','2027_affordability_difference']].sort_values(by='2027_affordability_difference')
```

```python
# List of suburbs to remove
suburbs_to_remove = ['Torquay', 'Highton', 'Geelong West - Hamlyn Heights', 'Geelong', 'North Geelong - Bell Park', 'Leopold', 'Newcomb']

tfouraffordability = tfouraffordability[~tfouraffordability['Suburb'].isin(suburbs_to_remove)]

# Filter out rows where 'Suburb' is in the list for tfiveaffordability
tfiveaffordability = tfiveaffordability[~tfiveaffordability['Suburb'].isin(suburbs_to_remove)]

# Filter out rows where 'Suburb' is in the list for tsixaffordability
tsixaffordability = tsixaffordability[~tsixaffordability['Suburb'].isin(suburbs_to_remove)]

# Filter out rows where 'Suburb' is in the list for tsevenaffordability
tsevenaffordability = tsevenaffordability[~tsevenaffordability['Suburb'].isin(suburbs_to_remove)]
```

```python
tfouraffordability.to_csv('data/landing/2024_affordability_rankings', index = False)
```
